## Basic Agent

Start by creating a simple agent that can answer questions and call tools. 

```python
#Load environment variables to store OPENAI_API_KEY
from dotenv import load_dotenv
import os   
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
# Crerate agent from langchain
from langchain.agents import create_agent
from pprint import pprint


def get_weather(city: str) -> str:
    """ Gets weather for a given city"""
    # Dummy function to simulate getting weather information
    return f"The weather in {city} is sunny with a high of 75°F."

agent = create_agent(
    model="gpt-4o",
    tools=[get_weather],
    system_prompt="You are a helpful assistant that provides weather information.",
)
#Prettyprint output

response = agent.invoke(
    {"messages": [{"role": "user", "content": "what is the weather in sf"}]}
)
```
