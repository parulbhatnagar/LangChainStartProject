## Requirements
Simple chatbot with memory using langchain


```python
#Load environment variables to store OPENAI_API_KEY
from dotenv import load_dotenv
import os   
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
# Take user input
user_input = input("You: ")
print(f"User input received: {user_input}")
```

```python
from langchain.chat_models import init_chat_model
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.messages import HumanMessage

model = init_chat_model(
    model="gpt-4o-mini",
    api_key=OPENAI_API_KEY,
    temperature=0.7
)

message_history = ChatMessageHistory()

# Do NOT set input_messages_key/history_messages_key
runnable_with_history = RunnableWithMessageHistory(
    model,
    lambda session_id: message_history
)

print("Chatbot: Hello! I'm a simple chatbot with memory. Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break

    # Pass a list of messages
    response = runnable_with_history.invoke(
        [HumanMessage(content=user_input)],
        config={"configurable": {"session_id": "user-1"}}
    )
    print(f"Chatbot: {response.content}\n")
```
